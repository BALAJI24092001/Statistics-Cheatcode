# Uniform / Rectangular Distribution

It is a probability distributions with approximately equally likely outcomes (in real-world dataset) within a given range.  

The uniform distribution encompasses both **discrete and continuous** variants.










## **Continuous Uniform Distribution**


*   In a continuous uniform distribution, outcomes are continuous and infinite.

  - **Example**: Heights range from 60 inches (5 feet) to 72 inches (6 feet).

### Probability Density Function (PDF):

The probability density function of a continuous uniform distribution is constant within a specified interval and zero elsewhere.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set the lower and upper bounds of the continuous uniform distribution
lower_bound = 5  # Lower bound in inches
upper_bound = 6  # Upper bound in inches

# Generate random ages from the continuous uniform distribution
num_people = 1000
ages = np.random.uniform(lower_bound, upper_bound, num_people)
```

```python
# Plot the histogram of ages
plt.figure(figsize=(8, 6))
plt.hist(ages, bins=30, density=True, color='orange', alpha=0.7)
plt.title('Histogram of Ages from Continuous Uniform Distribution')
plt.xlabel('Age (in inches)')
plt.ylabel('Density')
plt.show()
```

### Characteristics:

* Each value within the interval has an equal probability of occurrence.
* The PDF is a flat line, indicating a constant probability density within the interval.
* The cumulative distribution function (CDF) is a linear function, increasing uniformly from 0 to 1 within the interval.

## Can we plot QQ-Plot for Uniform Distribution?

Yes, a QQ plot is a scatterplot created by plotting two sets of quantiles against one another.

If both sets of quantiles came from the same distribution, we should see the points forming a line that's roughly straight.

Let's implement it practically.

```python
import scipy.stats as stats
import statsmodels.api as sm
```

```python
# Fit a uniform distribution to the data
params = stats.uniform.fit(ages)
distance = stats.uniform(loc=params[0], scale=params[1])
```

```python
# Create a QQ plot of the data using the uniform distribution
fig = sm.qqplot(ages, dist=distance, line='45')

# Add a title and labels to the plot
plt.title('QQ Plot of Uniform Distribution with Uniform Fit')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

# Show the plot
plt.show()
```

## **Discrete Uniform Distribution**

*   In a discrete uniform distribution, outcomes are discrete and have the same probability.

   - **Example**: Rolling of Dice.



### Probability Mass Function (PMF):

The probability mass function (PMF) of a discrete uniform distribution describes the likelihood of observing specific values in a finite set of equally likely outcomes.

It is defined as:

## **ùëì(ùë•)=1 / n‚Äã**

where,

  n,  total number of possible outcomes.






image.png

```python
# Simulate rolling a fair six-sided die
num_rolls = 1000
outcomes = np.random.randint(1, 7, size=num_rolls)

# Plot the outcomes
plt.figure(figsize=(8, 6))
plt.hist(outcomes, bins=np.arange(0.5, 7.5, 1), density=True, color='orange', alpha=0.7, rwidth=0.8)
plt.title('Histogram of Die Roll Outcomes')
plt.xlabel('Outcome')
plt.ylabel('Density')
plt.xticks(np.arange(1, 7, 1))
plt.show()
```

Wondering for Discrete Uniform Distribution?

Let's code!

```python
# Fit a uniform distribution to the data
params = stats.uniform.fit(outcomes)
dist = stats.uniform(loc=params[0], scale=params[1])
```

```python
# Create a QQ plot of the data using the uniform distribution
fig = sm.qqplot(outcomes, dist=dist, line='45')

# Add a title and labels to the plot
plt.title('QQ Plot of Discrete Uniform Distribution')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

# Show the plot
plt.show()
```

This alignment indicates that the observed frequencies of each outcome match the expected frequencies under the theoretical distribution, as per the Cumulative Distribution Function of Discrete uniform distribution.


### Applications of Uniform Distribution in ML and Data Science:



* **Random Initialization**: It is often used for randomly initializing the weights and biases of neural networks and other machine learning models, ensuring randomness and preventing biases during training.
* **Random Sampling**: It is utilized for random sampling techniques like random sampling without replacement and random shuffling of data during model training and evaluation.
* **Hyperparameter Tuning**: It is employed for sampling hyperparameter values from a uniform range during grid search or random search algorithms in hyperparameter tuning, helping in optimizing machine learning models.
* **Monte Carlo Simulation**: Uniform distribution serves as the basis for generating random numbers within a specified range in Monte Carlo simulations, which are used for modeling uncertainty and randomness in various scenarios such as finance, engineering, and risk analysis.
* **Random Forests**: In decision tree-based algorithms like Random Forests, uniform distribution is used for feature selection and splitting nodes, ensuring unbiasedness and randomness in the tree-building process.
* **Data Augmentation**: It can be used for generating random transformations such as rotations, translations, and scaling in data augmentation techniques for improving the diversity and size of training datasets.
